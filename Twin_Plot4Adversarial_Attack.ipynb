```python
# import pandas
import pandas as pd
#%matplotlib ipympl

import matplotlib.pyplot as plt
```

```python
# Carpentries link for gapminder data
data_url = '/content/Adversarial.csv'
#load gapminder data from url as pandas dataframe
gapminder = pd.read_csv(data_url)
print(gapminder.head(3))
```

```python
# create figure and axis objects with subplots()
fig,ax=plt.subplots()
ax.plot(gapminder.Types_of_Attack, gapminder.Attack_Success_Rate, marker="o")
ax.set_xlabel("Pixel Intensity")
ax.set_ylabel("Model Accuracy")
ax.plot(gapminder.Types_of_Attack, gapminder["Model_Accuracy"], marker="o")
plt.grid()
plt.show()
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
plt.grid()
# make a plot
ax.plot(gapminder.Types_of_Attack,
        gapminder.Model_Accuracy,
        color="red", 
        marker="o")
# set x-axis label
ax.set_xlabel("Types_of_Attack", fontsize = 12)
# set y-axis label
ax.set_ylabel("Model Accuracy",
              color="black",
              fontsize=12)
```

```python
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(gapminder.Types_of_Attack, gapminder["Attack_Success_Rate"],color="blue",marker="o")
ax2.set_ylabel("Attack Success Rate",color="blue",fontsize=12)
plt.show()
# save the plot as a file
fig.savefig('two_plot.pdf',
            format='pdf',
            dpi=100,
            bbox_inches='tight')
```
